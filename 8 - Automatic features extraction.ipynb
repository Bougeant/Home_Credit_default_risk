# Automatic features extraction from all dataset

This notebook creates automatic features from all auxilliary datasets by taking the average, min, max, sum and standard deviation of each relevant feature for the considered customer in the application dataset. These features add information regarding the customer that was not captured by the manually generated features in the previous notebooks.

```python
# Importing main modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import time
%matplotlib inline
pd.set_option('display.max_columns', 2000)

# Defining standard plot options
sns.set(font_scale=1.4)
sns.set_style("white")

# Reverting to old color palette
pal = sns.color_palette("deep")
pal = pal.as_hex()
pal[1], pal[2], pal[3] = pal[2], pal[3], pal[1]
sns.set_palette(pal)
```

```python
# Loading application dataframe from csv file 
application = pd.read_csv("./data/application.csv")
```

```python
# Function to create mean/min/max/sum/std features of other dataframes automatically
def automatic_feature_extraction(app, df, df_name, irrelevant_features = None):
    # Converting categorical features into dummy variables
    df = pd.get_dummies(df)
    
    # Removing irrelevent features 
    if irrelevant_features != None:
        relevant_features = df.drop(columns = irrelevant_features).columns
    else:
        relevant_features = df.columns
    # Generating automatic features
    # Mean of feature
    df_mean = df[relevant_features].groupby(by="SK_ID_CURR").mean().astype(np.float32)
    df_mean.columns = ["auto_" + str(df_name) + "_" + str(col) + '_mean' for col in df_mean.columns]
    df_mean.reset_index(inplace = True)
    app = pd.merge(app, df_mean, how = "left")
    
    # Min of feature
    df_min = df[relevant_features].groupby(by="SK_ID_CURR").min().astype(np.float32)
    df_min.columns = ["auto_" + str(df_name) + "_" + str(col) + '_min' for col in df_min.columns]
    df_min.reset_index(inplace = True)
    app = pd.merge(app, df_min, how = "left")
    
    # Max of feature
    df_max = df[relevant_features].groupby(by="SK_ID_CURR").max().astype(np.float32)
    df_max.columns = ["auto_" + str(df_name) + "_" + str(col) + '_max' for col in df_max.columns]
    df_max.reset_index(inplace = True)
    app = pd.merge(app, df_max, how = "left")

    # Sum of feature
    df_sum = df[relevant_features].groupby(by="SK_ID_CURR").sum().astype(np.float32)
    df_sum.columns = ["auto_" + str(df_name) + "_" + str(col) + '_sum' for col in df_sum.columns]
    df_sum.reset_index(inplace = True)
    app = pd.merge(app, df_sum, how = "left")

    # Standard deviation of feature
    df_std = df[relevant_features].groupby(by="SK_ID_CURR").std().astype(np.float32)
    df_std.columns = ["auto_" + str(df_name) + "_" + str(col) + '_std' for col in df_std.columns]
    df_std.reset_index(inplace = True)
    app = pd.merge(app, df_std, how = "left")

    return app
```

```python
# Generating automatic features for bureau dataset

bureau = pd.read_csv("./data/bureau.csv")
application = automatic_feature_extraction(application, bureau, "bureau", ["SK_ID_BUREAU"])
application.head()
```

```python
# Generating automatic features for bureau_balance dataset

# Merging bureau_balance and bureau dataframes to obtain link with application dataframe (SK_ID_CURR)
bureau_balance = pd.read_csv("./data/bureau_balance.csv")
bureau_balance = pd.merge(bureau_balance, bureau[["SK_ID_CURR", "SK_ID_BUREAU"]], on ="SK_ID_BUREAU", how = "right")
bureau_balance = bureau_balance.dropna(subset=['SK_ID_CURR'])
bureau_balance["SK_ID_CURR"] = bureau_balance["SK_ID_CURR"].astype(int)
application = automatic_feature_extraction(application, bureau_balance, "bureau_balance", ["SK_ID_BUREAU"])
application.head()
```

```python
# Generating automatic features for credit_card_balance dataset

credit_card_balance = pd.read_csv("./data/credit_card_balance.csv")
application = automatic_feature_extraction(application, credit_card_balance, "credit_card_balance", ["SK_ID_PREV"])
application.head()
```

```python
# Generating automatic features for pos_cash_balance dataset

pos_cash_balance = pd.read_csv("./data/POS_CASH_balance.csv")
application = automatic_feature_extraction(application, pos_cash_balance, "pos_cash_balance", ["SK_ID_PREV"])
application.head()
```

```python
# Generating automatic features for previous_application dataset

previous_application = pd.read_csv("./data/previous_application.csv")
# Some features are removed because they would lead to a huge number of dummy variables (and to memory issues) 
irrelevant_features = ["SK_ID_PREV", "NAME_CASH_LOAN_PURPOSE", "WEEKDAY_APPR_PROCESS_START", "NAME_GOODS_CATEGORY", "PRODUCT_COMBINATION"]
previous_application = previous_application.drop(columns = irrelevant_features)
application = automatic_feature_extraction(application, previous_application, "previous_application")
application.head()
```

```python
# Generating automatic features for installments_payments dataset

installments_payments = pd.read_csv("./data/installments_payments.csv")
application = automatic_feature_extraction(application, installments_payments, "installments_payments", ["SK_ID_PREV"])
application.head()
```

```python
# Saving application dataframe into csv file
application.to_csv("./data/application.csv",index=False)
```
